Imports

```python
import requests
import pandas as pd
import Keystuff
import json
import time
```

Main

```python
log_df = pd.read_csv(r"~/CoinStuff/coin_cap_data.csv")


url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    
parameters = {
        'start': '1',
        'limit': '25',
        'convert': 'USD'
    }
    
header_stuff = {
        'Accepts': 'application/json',
        'X-CMC_PRO_API_KEY': Keystuff.APIKey,
    }
    
resp = requests.get(url, params=parameters, headers=header_stuff)
   
json_data = json.loads(resp.text)
    
coin_df = pd.json_normalize(json_data['data'])
    
final_df = log_df.append(coin_df, ignore_index=True)
    
final_df.to_csv(r'~/Projects/coin_cap_data.csv')

print("Successful Run")
```
